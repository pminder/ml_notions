# Les courbes ROC, une fois pour toutes

Ce notebook propose une introduction à la notion de courbe ROC (Receiver Operating Characteristic, en hommage à leur origine militaire). L'idée est de comprendre leur fonctionnement, la façon dont elle sont construites ainsi que d'apprendre à les lire et à les utiliser. Aucune formalisation mathématique en perspective. Le lecteur intéressé est bien évidemment invité à se réferrer aux annexes pour plus de détails ;)

## Introduction

L'idée sous-jacente à la notion de courbe ROC est de se servir de la nature probabiliste des modèles que l'on utilise parfois - voire souvent - en machine learning. Contrairement à d'autres mesures telles que les matrices de confusion, il ne suffit pas pour tracer une courbe ROC de disposer du résultat de la classification d'une ou plusieurs observations par un modèle. On veut aller plus loin et, en quelque sorte, plonger dans le détail de l'obtention de ce résultat. Evidemment, cela nécessite une compréhension du modèle en tant soit peu plus avancée (mais, que le lecteur paresseux se rassure, nombreuses sont les bibliothèques qui nous épargnent en fait le gros du travail). 

Pour fixer les idées, plaçons dans le cadre le plus courant en matière de courbes ROC, celui d'un indicateur binaire. A partir d'un ensemble de variables, on souhaite prédire la classe d'une ou plusieurs observations, appartenant nécessairement à l'ensemble $\{0, 1\}$. Pour tester la qualité de notre modèle, on dispose de nombreuses mesures. On pourrait par exemple calculer le nombre d'observations bien classées sur le nombre d'observations total. C'est ce que l'on appelle la précision du modèle. On pourrait aussi, pour la classe 0 et la classe 1, compter le nombres d'observations bien classées et mal classées et présenter le tout sous la forme d'un tableau. C'est ce que l'on appelle une matrice de confusion. On pourrait aussi calculer d'autres indicateurs tels que le rappel ou la 'sensitivité'... Toutes ces mesures ont en commun le fait qu'elle nécessitent seulement de connaître les résultats de la classification ainsi que la classe véritable des observations. Cela peut s'avérer pratique dans le cas d'un modèle boîte noire ou nous ne disposons justement pas d'autres informations. 

Mais supposons maintenant que nous disposions non pas d'une simple réponse binaire de la part du modèle (du type 'cette observation appartient à la classe 1, fais-moi confiance, je sais ce que je dis') mais bien plutôt de la probabilité d'appartenance de l'observation en question à chacune des classes. Cette information supplémentaire est loin d'être évidente et n'a rien de trivial a priori. C'est en ce sens qu'il importe de bien connaître son modèle et de s'assurer de la pertinence de cette probabilité. Certains modèles comme NaiveBayes reposent en fait intégralement sur le calcul de telles probabilités. D'autres, comme les SVM, peuvent être très légèrement adaptés afin d'obtenir cette information supplémentaire. Nous ne discuterons pas ici de la manière d'obtenir cette probabilité car ce n'est pas vraiment notre propos. Supposons simplement que nous disposions d'un modèle capable de nous la fournir. Et voyons comment les courbes ROC nous permettent d'exploiter cette information supplémentaire.

## Un modèle simplifié pour bien comprendre

Commençons sans plus attendre par générer quelques observations et construisons quelques petits modèles que nous pourrons ensuite comparer à l'aide des courbes ROC. 

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

Nous générons des points dans le disque de rayon 1 et leur attribuons le label 1 avec une probabilité d'autant plus forte qu'ils sont éloignés de l'origine du plan. Plus précisément, les points de label 0 suivent (presque) une loi normale centrée réduite dans $\mathbb{R}^2$. Les points de labels sont obtenus à patrtir de leur coordonnées polaires:
- $\theta$ suit une loi uniforme sur $[0, 2\pi]$
- $r$ suit une loi normale $\mathcal{N}(3, 1)$

On génère 1000 points pour chaque label et on normalise à 1 le point le plus éloigné de l'origine.

```python
N = 1000

reds = .9*np.random.normal(size = (N, 2))

rblues = 3 - np.random.normal(size = N)
tblues = 2*np.pi*np.random.random(size = N)
blues = np.array([rblues*np.cos(tblues), rblues*np.sin(tblues)]).T

M = np.max(blues)
reds = reds/M
blues = blues/M
```

Un petit graphique pour fixer les idées

```python
plt.figure(figsize = (15, 5))

plt.subplot(1, 2, 1)
plt.scatter(reds[:,0], reds[:,1], c = 'r')
plt.scatter(blues[:,0], blues[:,1], c = 'b')
plt.axis([-1, 1, -1, 1]);

plt.subplot(1, 2, 2)
plt.scatter(reds[:,0], reds[:,1], c = 'r')
plt.scatter(blues[:,0], blues[:,1], c = 'b')
plt.axis([-.5, .5, -.5, .5]);
```

On rassemble tous ces points et leur labels dans les traditionnelles matrices $X$ et $Y$.

```python
X = np.concatenate([blues, reds])
Y = np.concatenate([np.ones(N), np.zeros(N)]).T
print X.shape
print Y.shape
```

#### Création des modèles

Voici les modèle que nous nous proppsons d'étudier et de comparer:
- le premier attribue le label 1 à un point $x = (x_1, x_2)$ avec une probabilté $\sqrt{||x||_{+\infty}} = \sqrt{\max(|x_1 |, |x_2|)}$
- le second attribue le label 1 à un point situé situé à une distance $r$ de l'origine avec une probabilité $\sqrt{r}$ 
Comme nous le verrons plus loin, ces modèles sont les plus proches de la réalité. Mais, pour les besoins de la cause, voici deux modèles que nous essaierons de comparer entre eux:
- le troisième modèle est très similaire au second mais beaucoup moins bien calibré. Il attribue le label 1 à un point situé situé à une distance $r$ de l'origine avec une probabilité $r$ 
- le quatrième modèle est plus saugrenu. Il attribue le label 1 à un point $x = (x_1, x_2)$ avec une probabilté $\min(|x_1|, |x_2|)$

Remarque: j'ai un peu triché en ajoutant quelques puissances par-ci par-là pour rendre les résultats plus visuels et plus probants. Quoi qu'il en soit, le raisonnement reste le même et les conclusions sont inchangées (simplement moins flagrantes).

```python
def model1(X):
    """Norme infinie"""
    return np.max(np.abs(X), axis = 1)**.5

def model2(X):
    """Norme euclidienne"""
    R = np.sqrt(X[:,0]**2 + X[:,1]**2)
    return R**.5

def model3(X):
    """Norme euclidinne mal calibrée"""
    R = np.sqrt(X[:,0]**2 + X[:,1]**2)
    return R**1.15

def model4(X):
    """Norme ésotérique qui porte peut-être un nom que je ne connais pas"""
    return np.min(np.abs(X), axis = 1)**.3
```

## Comparaison des modèles à l'aide de différentes mesures

On commence par générer les prédictions de ces trois modèles: on choisit la classe la plus probable, en fonction des probabilités renvoyées par les modèles.

```python
Y_1 = np.round(model1(X)).astype('bool')
Y_2 = np.round(model2(X)).astype('bool')
Y_3 = np.round(model3(X)).astype('bool')
Y_4 = np.round(model4(X)).astype('bool')

plt.figure(figsize=(20, 5))
plt.subplot(141)
plt.scatter(X[Y_1][:,0], X[Y_1][:,1])
plt.scatter(X[~Y_1][:,0], X[~Y_1][:,1], c = 'r')
plt.axis([-.5, .5, -.5, .5]);

plt.subplot(142)
plt.scatter(X[Y_2][:,0], X[Y_2][:,1])
plt.scatter(X[~Y_2][:,0], X[~Y_2][:,1], c = 'r')
plt.axis([-.5, .5, -.5, .5]);

plt.subplot(143)
plt.scatter(X[Y_3][:,0], X[Y_3][:,1])
plt.scatter(X[~Y_3][:,0], X[~Y_3][:,1], c = 'r')
plt.axis([-.5, .5, -.5, .5]);

plt.subplot(144)
plt.scatter(X[Y_4][:,0], X[Y_4][:,1])
plt.scatter(X[~Y_4][:,0], X[~Y_4][:,1], c = 'r')
plt.axis([-.5, .5, -.5, .5]);
```

Ces graphiques nous permettent de nous faire une bonne idée de la façon dont nos quatre modèles reflètent la réalité. Les deux premiers modèles semblent assez bons. Mais, d'un point de vue théorique, on aura tendance à préferer le second puisque la forme géométrique associée à la norme eucliedienne (un disque et non un carré) est plus proche de la façon dont nous avons généré nos points (à l'aide de lois normales, ce qui fait que les points sont organisés en cercle autour de l'origine).

Les deux derniers modèles sont beaucoup plus mauvais. Le troisième attribue beaucoup trop de labels 0 (points rouges) même si la forme des prédictions est satisfaisante. Quant au quatrième, le critère utilisé est vraisemblablement absurde. La qualité de ses prédictions en sera sans doute d'autant plus mauvaise.

### Mesure n°1 : la précision

On calcule le nombre d'erreurs commises par nos trois modèles.

```python
n1 = (Y != Y_1).sum()
n2 = (Y != Y_2).sum()
n3 = (Y != Y_3).sum()
n4 = (Y != Y_4).sum()
plt.bar([.1, 1, 1.9, 2.8], [n1, n2, n3, n4])
```

Deux conclusions s'imposent si l'on s'en tient à cette mesure:
- tout d'abord, on ne peut pas vraiment distinguer le modèle 3 du modèle 4 en terme de nombre d'erreurs commises. Les deux modèles (pourtant très différents) ont des performances similaires (le modèle 3 semble un peu meilleur mais cela n'est pas flagrant). Cela n'est pas satisfaisant du tout! Nous savons que le modèle 3 est bien meilleur que le modèle 4. On voit là une insuffisance du simple calcul du nombre d'erreurs commises par un modèle pour juger de sa qualité. Les courbes ROC, nous le verrons nous permettent en partie de pallier ce problème. 
- ensuite, les modèles 1 et 2 sont assez proches et beaucoup plus performants que les deux autres. On note cependant une légère supériorité du modèle 1 sur le modèle 2. Là encore, connaissant la façon dont nous avons généré nos points, on a le droit de rester dubitatif...

### Mesure n°2 : la courbe ROC (et oui...)

L'idée est donc de se servir des probabilités renvoyées par le modèle et non simplement de la classe attribuée par ce dernier à nos observations. Pour ce faire, on commence par fixer un seuil $s$. A partir de là, on attribue à chaque observation $i$ (en notant $p_i$ la probabilité associée à l'observation $i$):
- le label 1 si $p_i > s$
- le label 0 si $p_i \leq s$

En fait, nous avons procédé exactement de la même façon mais en fixant implicitement $s = 0,5$. Tout l'intérêt de la démarche consiste à faire cette opération pour un $s$ quelconque, que l'on fera ensuite varier dans $[0, 1]$.

Ensuite, on calcule:
- le "True Positive Rate" (ou TPR): la fraction d'oservations appartenant à la classe 1 et auxquelles on a attribué le label 1
- le "False Positive Rate" (ou FPR): la fraction d'obsevations appartenant à la classe 0 et auxquelles on a attribué le label 1

Les noms anglais sont un peu trompeurs et portent facilement à confusion (on a inventé plein de dénominations différentes pour désigner la même chose et on désigne souvent de la même façon plusieurs réalités différentes). Donc mieux vaut ne pas trop se préoccuper des noms et se concentrer sur les notations et les définitions.

En faisant varier $s$, on obtient un certain nombre de couples $(FPS(s), TPS(s))$. Une courbe ROC n'est rien d'autre que la courbe obtenue de cette façon, en plaçant FPS en abscisses et TPS en ordonnées. Et oui, c'est aussi simple que cela. 

Voyons, sans plus attendre ce que cela donne pour notre dernier modèle.

```python
FPR, TPR, _ = roc_curve(Y, model4(X))
plt.plot(FPR, TPR)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```

On peut remarquer plusieurs choses sur le graphique ci-dessus:
- pour $s=0$, on a pour tout $i$, $p_i \geq s$. Toutes les observations se voient donc affublées du label 1. La fraction d'observations appartenant à la classe 1 et auxquelles on a attribué le label 1 est donc égale à 1 ($TPR = 1$). Et de même, la fraction d'observations appartenant à la classe 0 et auxquelles on a attribué le label 1 est égale à 1 ($FPR = 1$). On obtient ainsi le point $(1, 1)$ de la courbe.
- de même lorsque $s=1$, on obtient le point de la courbe situé en $(0, 0)$ (mais ce point apparaît moins clairement sur le graphique)
- on peut aussi se convaincre qu'un modèle qui attribuerait la même probabilité à chaque observation (par exemple la fréquence d'apparition de la classe 1 dans l'échantillon total) aurait pour courbe ROC la droite passant par $(0, 0)$ et $(1, 1)$. En fait, ces deux points seraient les seuls que l'on obtiendrait. En notant $p$ cette unique probabilité, on aurait:

$$
\begin{cases}
\forall s \in [0, p],& FPR = TPR = 0 \\
\forall s \in \; ]p, 1],& FPR = TPR = 1
\end{cases}
$$

L'idée est donc de mesurer la qualité d'un modèle en fonction de la façon dont il s'éloigne de la droite diagonale (celle passant par l'origine et de pente 1). Plus la courbe ROC d'un modèle s'éloigne de cette droite et se rapproche de celle passant par le point $(0, 1)$ où toutes les prévisions sont correctes et plus le modèle est pertinent (étant donnée la base sur laquelle on le test). 

Por s'en convaincre, deux graphiques, celui de gauche représente un modèle à probabilité constante, celui de droite un modèle parfait.

```python
plt.figure(figsize=(15, 5))

FPR, TPR, _ = roc_curve(Y, .2*np.ones_like(Y))
plt.subplot(121)
plt.plot(FPR, TPR)
plt.title(u"Un modèle à probabilité constante")

FPR, TPR, _ = roc_curve(Y, Y)
plt.subplot(122)
plt.axis([0, 1, 0, 1.01])
plt.plot(FPR, TPR)
plt.title(u"Un modèle idyllique")
```

Dans ce contexte, on choisit comme indicateur synthétique de la qualité de notre modèle l'aire située sous la courbe ROC. Plus celle-ci est proche de 1 et meilleur est le modèle. Si jamais cette aire s'avère inférieure à 0.5, c'est que le modèle est très mauvais et peut être rejetté aussitôt (il suffit pour faire au moins aussi bien de prédire systématiquement la classe la plus nombreuse au sein de l'échantillon).

Forts de ces quelques considérations, nous proposons d'utiliser les courbes ROC pour comparer les modèles 3 et 4.

```python
fig, ax = plt.subplots()

FPR, TPR, _ = roc_curve(Y, model3(X))
area = auc(FPR, TPR)
ax.plot(FPR, TPR, label = u'Modèle 3 (aire = {})'.format(round(area, 3)))

FPR, TPR, _ = roc_curve(Y, model4(X))
area = auc(FPR, TPR)
ax.plot(FPR, TPR, label = u'Modèle 4 (aire = {})'.format(round(area, 3)))

plt.plot([0, 1], [0, 1], '--')
ax.legend(loc = 'lower right')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
```

Aha!! Nous sommes parvenus à faire apparaître très clairement la différence de qualité qui sépare ces deux modèles! Le modèle 4 est, en terme de courbe ROC bien moins performant que le modèle 3. Cela vient du fait que les courbes ROC jugent davantage de l'architecture du modèle, de la façon dont celui-ci produit ses prévisions plutôt qu'aux prévisions en elle-mêmes. Le seuil $s$ nécessaire pour atteindre une même fraction de points classés 1 parmi les points appartenant effectivement à la classe 1 est beaucoup plus faible dans le cas du modèle 4 que du modèle 3. Cela signifie que, dans le cas du modèle 4, on attribue la classe 1 à beaucoup plus de points. La fraction de points appartenant à la classe 0 et auxquels on a attribué le label 1 est donc nettement plus importante pour le modèle 4 que pour le modèle 3, d'où un FPR bien plus élevé.

Continuons sur notre lancée et traçons les courbes ROC associées aux modèles 1 et 2.

```python
fig, ax = plt.subplots()

FPR, TPR, _ = roc_curve(Y, model1(X))
A1 = auc(FPR, TPR)
ax.plot(FPR, TPR)

FPR, TPR, _ = roc_curve(Y, model2(X))
A2 = auc(FPR, TPR)
ax.plot(FPR, TPR)

plt.plot([0, 1], [0, 1], '--')
ax.legend((u'Modèle 1 (aire = {})'.format(round(A1, 3)),
        u'Modèle 2 (aire = {})'.format(round(A2, 3))),
        loc = 'lower right');

ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
```

Cette fois-ci la différence est beaucoup moins nette, ce qui n'est pas absurde: les modèles 1 et 2 utilisent tous les deux une distance assez classique pour formuler leurs prévisions. Mais contrairement à ce que semblait indiquer la mesure de précision, les courbes ROC semblent donner l'avantage au modèle 2, ce qui est tout à fait cohérent avec la façon dont nous avons généré nos points! En réalité, je ne crois pas que, dans une situation réelle, une si petite différence suffirait pour privilégier le premier modèle par rapport au second. Mais elle permet tout de même assurément de nuancer les conclusions que l'on pourrait tirer en s'appuyant sur d'autres mesures et suffit pour mettre la puce à l'oreille du statisticien avisé ;) On pourrait par exemple se dire que le modèle 2 a un meilleur potentiel que le modèle 1 et que c'est donc lui qu'on a intérêt à calibrer en premier...

Une dernière remarque pour nuancer cette fois-ci l'approche par courbes ROC: les aires associée aux modèles 2 et 3 sont exactement les mêmes! Les courbes ROC ne comparent pas les performances du modèle en tant que telles mais plutôt la façon dont les probabilités d'appartenance à une classe plutôt qu'une autre sont hiérarchisées. Dans le cas des modèles 2 et 3, les probabilités d'appartenance sont hiérarchisées de la même façon. En fait ces deux modèles sont très proches (on a simplement ajouté un carré). Leur fonctionnement interne est donc en quelque sorte similaire et les courbes ROC ne voient aucune différence entre les deux. C'est au niveau des résultats de classification que la différence se manifeste (et force est de constater que dans notre cas, elle n'est pas des moindres!).

### Conclusion

On peut retenir de cette petite introduction que les courbes ROC sont des mesures de performance de modèles statistiques qui peuvent s'avérer utiles pour mettre en évidence le bon ou le mauvais fonctionnenment d'un modèle en termes de principe de fonctionnement. Notre dernière remarque nous invite cependant à nous rappeler que les courbes ROC ne sont évidemment pas parfaites et qu'elles gagnent non pas à être utlisées à la place mais bien en complément d'autres mesures de performances.
